```python
import random
import pyshark
from scapy.all import *
```

```python
ports = []
for i in range(2**10+1,2**16):
    if random.random() <= 0.95:
        ports.append(i)
```

```python
capture = pyshark.LiveCapture(interface='wlp3s0', bpf_filter='tcp')
```

```python
capture.sniff(packet_count=10)
```

```python
def syn_reply(pkt):
    ipdst = str(pkt.layers[1]._all_fields['ip.dst'])
    ipsrc = str(pkt.layers[1]._all_fields['ip.src'])
    portdst = int(pkt.layers[2]._all_fields['tcp.dstport'])
    portsrc = int(pkt.layers[2]._all_fields['tcp.port'])
    seq = int(pkt.layers[2]._all_fields['tcp.seq'])+1
    
    return IP(src=ipdst, dst=ipsrc)/TCP(sport=portdst, dport=portsrc)
```

```python
for pkt in capture:
    if int(pkt.layers[2]._all_fields['tcp.flags.syn']) == 1:
        for closed in ports:
            if closed == int(pkt.layers[2]._all_fields['tcp.dstport']) and str(pkt.layers[1]._all_fields['ip.src']) != '10.25.134.237':
                print(pkt)
                #send(syn_reply(pkt))
```
